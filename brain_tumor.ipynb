```
t_v = np.zeros((253,2),dtype=int)
```

```
for i in range(len(training_values)):
  if training_values[i] == 1:
    t_v[i][0] = 0
    t_v[i][1] = 1
  elif training_values[i] == 0:
    t_v[i][0] = 1
    t_v[i][1] = 0
```

```
X_train,X_test,Y_train,Y_test = train_test_split(training_images,t_v,test_size=0.1)
```

# CNN model

```
model=Sequential()
model.add(Conv2D(32,(3,3),activation='relu',input_shape=(150,150,3)))
model.add(MaxPooling2D((2,2)))

model.add(Conv2D(64,(3,3),activation='relu'))
model.add(MaxPooling2D(2,2))
model.add(Conv2D(128,(3,3),activation='relu'))
model.add(MaxPooling2D(2,2))

model.add(Flatten())
model.add(Dense(2,activation='softmax'))
```

```
model.summary()
```

```
model.compile(optimizer='adam', loss='categorical_crossentropy',metrics=['accuracy'])
```

```
hist = model.fit(X_train,Y_train,validation_data=(X_test,Y_test),epochs=15,batch_size=32)
```
